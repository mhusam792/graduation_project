```python
from ultralytics import YOLO
import torch
from PIL import Image
```

```python
def is_image(file_path):
    """
    Check if the given file is a valid image.

    Parameters:
    - file_path (str): The path to the file.

    Returns:
    - bool: True if the file is a valid image, False otherwise.
    """
    try:
        # Attempt to open the image file
        with Image.open(file_path):
            return True
    except (IOError, OSError, Image.DecompressionBombError):
        # IOError and OSError are raised if the file cannot be opened or is not an image
        # Image.DecompressionBombError is raised if the image is too large
        return False

def model_result(model_path, img_path):
    # Checking if we are using "GPU" or "CPU"
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print('Using Device: ', device)

    # Load a pretrained model
    model = YOLO(model_path)

    #results = model.predict(img_path, save=True, imgsz=320, conf=0.5) # لو عايز تعمل حفظ للصوره
    results = model.predict(img_path, device=device, save=True, conf=.5)
    return results, model

def return_cls(modle_result):
    '''Functiion that retrieves classes in image'''
    clist= results[0].boxes.cls
    cls = list()
    for cno in clist:
        cls.append(model.names[int(cno)])
    return cls

def count_cls(input_list):
    '''A function that retrieves the number of times that each class has appeared in image'''
    label_counts = {}
    
    for label in input_list:
        if label in label_counts:
            label_counts[label] += 1
        else:
            label_counts[label] = 1
    
    return label_counts
```
