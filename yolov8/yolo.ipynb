```python
from ultralytics import YOLO
import torch
from PIL import Image
import os
from shutil import copy2, rmtree
import cv2
```

```python
def is_image(file_path):
    """
    Check if the given file is a valid image.

    Parameters:
    - file_path (str): The path to the file.

    Returns:
    - bool: True if the file is a valid image, False otherwise.
    """
    try:
        # Attempt to open the image file
        with Image.open(file_path):
            return True
    except (IOError, OSError, Image.DecompressionBombError):
        # IOError and OSError are raised if the file cannot be opened or is not an image
        # Image.DecompressionBombError is raised if the image is too large
        return False

def model_result(model_path, img_path):
    """
    Result of Model prediction.

    Parameters:
    - model_path (str): path of specific trained model.
    - img_path (str): path of image you want to predict.

    Returns:
    - results: Result of predicted image result.
    - model: to can use all functions that developed by ultralytics it self. 
    """
    # Checking if we are using "GPU" or "CPU"
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print('Using Device: ', device)

    # Load a pretrained model
    model = YOLO(model_path)

    #results = model.predict(img_path, save=True, imgsz=320, conf=0.5) # لو عايز تعمل حفظ للصوره
    results = model.predict(img_path, device=device, save=True, project="runs/detect/prediction")
    return results, model

def return_cls(modle_result, trained_model):
    """
    Functiion that retrieves classes in image.

    Parameters:
    model_result: The result of model prediction in specific image.
    trained_model: The be Trained model you have loaded.

    Returns:
    A list of all classes that have appeared in specific image.
    """
    clist= modle_result[0].boxes.cls
    cls = list()
    for cno in clist:
        cls.append(trained_model.names[int(cno)])
    return cls

def count_cls(input_list):
    """
    A function that retrieves the number of times that each class has appeared in image.

    Parameters:
    - input_list (list): list of all classes that have appeared in the image.

    Returns:
    - A Dictionary of items as a Key and number of each item have appeared in image as a Value.
    """
    label_counts = {}
    
    for label in input_list:
        if label in label_counts:
            label_counts[label] += 1
        else:
            label_counts[label] = 1
    
    return label_counts

def copy_images_and_remove_folder(source_folder, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Get the list of files in the source folder
    files = os.listdir(source_folder)

    for file_name in files:
        # Construct the full path of the source and destination files
        source_path = os.path.join(source_folder, file_name)
        destination_path = os.path.join(destination_folder, file_name)

        # Copy the image file to the destination folder
        copy2(source_path, destination_path)

    # Remove the source folder
    rmtree(source_folder)

def display_image(image_path):
    # Read the image using OpenCV
    image = cv2.imread(image_path)

    # Display the image
    cv2.imshow('Image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
